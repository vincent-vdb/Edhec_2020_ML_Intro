# II. Build great visualizations

## II.1. Choose _carefully_ your chart - what is your message?

Data visualization is all about **storytelling**: what message do you want to convey?

An histogram will highlight something different than a scatter plot, that will tell a different story than a line chart, etc.

Start by answering some questions about your data and the goal of your visualization:
- Are you trying to answer a question (ex: are the iPhone sales growing over time - at what pace)? To confirm/infirm an hypothesis (ex: is global warming real)? To convey a message (ex: should you buy Bitcoin **now**)?
- What would be the best chart(s) to do so?
- What data do you need to draw this/these chart(s)?
- Collect this data (you know how to do that now), aggregate it
- Draw your chart(s)! üé®

## II.2. Maximize the data-ink ratio 

The **Data-Ink ratio** is a concept introduced by Edward Tufte, the expert whose work has contributed significantly to designing effective data visualizations.

In his 1983 book, *The Visual Display of Quantitative Data*, he stated the goal:
> Above all else, show the data.
>
> *Tufte, 1983*

He then defined the **data-ink ratio** as **the proportion of ink** (or pixels) **that is used to present actual data compared to the total amount of ink** (or pixels) **used in the entire display**.

$$Data \: Ink \: Ratio = \frac{Data \: Ink}{Total \: ink \: used \: in \: graphic}$$

In other words, focus on showing the data and remove everything irrelevant.

As we say: **LESS IS MORE**.

#### Decorating your plot

You can then detail properly:
- the title of the figure: `plt.title('Chart Title')`
- the axis label: with `plt.xlabel("X-title")` and `plt.ylabel("Y-title")`

As you will see in the documentation, matplotlib offers a variety of parameters in these functions to custom your charts as you wish.

After all that, you call the function `plt.show()` in order to draw the figure.

````python
# Creating the figure
plt.figure()

# We can add axis labels
plt.xlabel('X-axis (in units)', fontsize='14')
plt.ylabel('Y-axis (in units)', fontsize='14')

# We can add and format a title for our chart
plt.title('My wonderful chart', 
          fontweight='bold', color='black', fontsize='16', alpha=0.8)

# At the end, we display the plot with `plt.show()`
plt.show()

# It only displays the axis here because we did not specify a plot to draw nor values to take
````

#### `plt.figure`

When creating a chart with `pyplot`, it creates automatically a figure (container for your charts).

But you can create it yourself in order to, for example, specify the size of your figure with `figsize=(width, height)` with dimensions in inches.

````python
# Same chart as before, but first we create a figure (container for your charts)
# with an appropriate size
plt.figure(figsize=(12,6))

# We can add axis labels
plt.xlabel('X-axis (in units)', fontsize='14')
plt.ylabel('Y-axis (in units)', fontsize='14')

# We can add and format a title for our chart
plt.title('My wonderful chart', 
          fontweight='bold', color = 'black', fontsize='16', alpha=0.8)

# At the end, we display the plot with `plt.show()`
plt.show()
````

#### `plt.hist`

You can easily draw histograms with `pyplot` using the `plt.hist` function.

This allows you to visualize the **distribution** of your data into different **bins** (or bars or buckets) of the same size.

Bins represents identical range of values, the height of the bar corresponds to the number of values contained in this range.

````python
# Distribution > Histogram (column or line)
import numpy as np
mu = 200
sigma = 25
x = np.random.normal(mu, sigma, size=1000)

plt.figure(figsize=(10,5))

# Here we specify that we want 20 bins for our histogram
plt.hist(x, bins=20, facecolor='blue', alpha=0.5)

plt.title('Histogram of Normal Distribution (random draw)', fontweight='bold')
plt.show()
````

#### `plt.scatter`

You can visualize data points on a X vs Y axis by using `plt.scatter`

It is very useful when looking for **correlations** between 2 variables for example.

This function draws a scatter plot of the observations. It displays values for typically two variables for a set of data. You can also display more variables by coloring the points, or varying the size.

````python
# Relationship > 2D > Scatter chart
x = range(1,101)
y = x + np.random.randn(100) * 15

plt.figure(figsize=(10,5))

plt.scatter(x=x, 
            y=y, 
            marker='o',
            alpha=0.5)

plt.show()
````

````python
# Relationship > 3D > Bubble chart
# It is a scatter plot with varying size (variable `s`) of the points
x = np.random.rand(100) - 0.5
y = x + np.random.rand(100)
z = x + x * np.random.rand(100)

plt.figure(figsize=(10,5))

plt.scatter(x, y, s=z * 1000, alpha=0.5)

plt.show()
````

````python
# Relationship > 3D > Scatter colored chart (discrete or continue variables)
# It is a scatter plot with varying color (variable `c`) depending on the points

# Loading iris dataset - we will talk about it later
import seaborn as sns    # seaborn is another very useful data visualization library
import pandas as pd
df = sns.load_dataset('iris')
df['species'] = pd.Categorical(df['species'])

# Constructing the figure
plt.figure(figsize=(10,5))

plt.scatter(x=df['sepal_length'],
            y=df['sepal_width'],
            c=df['species'].cat.codes,
            cmap="Set1", # `cmap` is the colormap (or palette), try "rainbow" for example
                         # List of plt colormaps and how to choose it:
                         # https://matplotlib.org/examples/color/colormaps_reference.html
            alpha=0.8)   # alpha is transparency parameter

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width') 
plt.show()
````

#### `plt.plot`

One of the most classical chart: `plt.plot`.

Draws lines between the data points. Especially useful with time series.

````python
# Evolution > Many data points (ex: Time series) > Line chart

plt.figure(figsize=(10,5))

x = range(1900, 2000)
y = range(100) + np.random.randn(100) * 15

plt.plot(x, y,
         marker='x', # Markers used for your data points - try 'o'
         linestyle='--', # Style used for your lines - try '-'
         color='mediumvioletred',
         alpha=0.5)

plt.title("Evolution of xxx during 21th century")
plt.show()
````

#### `plt.bar`

You can build a **column chart** with `plt.bar`.

It is mostly used to compare data with a few items of few categories.

````python
# Evolution > Few data points > Column chart
plt.figure(figsize=(10, 5))

bars = ['Q1 2018', 'Q2 2018', 'Q3 2018', 'Q4 2018']
sales = [12, 14, 25, 59]

plt.bar(x=bars,
        height=sales)

plt.ylabel("Sales (in M‚Ç¨)")
plt.title("Sales have skyrocketed Month-Over-Month (MoM)!", 
          fontweight='bold', fontsize='14', alpha=0.8)
plt.show()
````

#### `plt.stackplot`

If you want to observe the composition of some data and its **relative** evolution over time, you can use the `plt.stackplot` function.

This will plot and color the evolution of your groups share (max 100%).

````python
# Evolution Share Of Voice (SoV) > Stacked 100% column chart

# Creating the data
data = pd.DataFrame({'group_A':[1,4,6,8,9],
                     'group_B':[2,24,7,10,12],
                     'group_C':[2,8,5,10,6], },
                     index=range(1,6)
                   )
# We need to transform the data from raw data to percentage (fraction)
data_perc = data.divide(data.sum(axis=1), axis=0)

# Creating the figure
plt.figure(figsize=(10, 5))

# Make the plot
plt.stackplot(range(1,6),
              data_perc["group_A"], 
              data_perc["group_B"], 
              data_perc["group_C"],
              labels=['A','B','C'])

# Formatting the plot
plt.legend(loc='upper left')
plt.margins(0,0)
plt.title('Evolution of market share over time')
plt.show()
````

#### `plt.subplot` - digging deeper

Now let's demo briefly how to showcase multiple charts at the same time with `plt.subplot`

It works likes this: `plt.subplot(nrows, ncolumns, index)`


<p align="center">

</p>

You can use `plt.tight_layout()` to automatically arrange your plots (and prevent overalapping for example)

````python
# Loading data
x = range(1,101)
y = np.random.randn(100) * 15 + range(1,101)
z = (np.random.randn(100) * 15 + range(1,101)) * 2
````

````python
# Display charts in two columns
 
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.plot( x, y, marker='o', alpha=0.4)

plt.subplot(1, 2, 2)
plt.plot(x, y, linestyle='none', marker='o', color="orange", alpha=0.5)

plt.tight_layout()
plt.show()
````

````python
# Display charts in two rows - Note that we can get rid of the "," in the subplot function
 
plt.figure(figsize=(10, 8))

plt.subplot(211)
plt.plot(x, y, marker='o', alpha=0.4)
 
plt.subplot(212)
plt.plot(x, y, linestyle='none', marker='o', color="orange", alpha=0.5)

plt.tight_layout()
plt.show()
````

You can even do more advanced stuff with `plt.subplot2grid`. For example:

````python
# More advanced/custom charts arrangement with `plt.subplot2grid`
# Example: grid of 4 columns and 2 rows
 
plt.figure(figsize=(12,10))

# The first plot is on line 1, and is spread all along the 4 columns
ax1 = plt.subplot2grid((2, 4), (0, 0), colspan=4)
ax1.plot(x, y , marker='o', alpha=0.5)

# The second one is on column2, spread on 3 columns
ax2 = plt.subplot2grid((2, 4), (1, 0), colspan=3)
ax2.plot(x, y, marker='o', color="grey", alpha=0.5)

# The last one is spread on 1 column only, on the 4th column of the second line.
ax3 = plt.subplot2grid((2, 4), (1, 3), colspan=1)
ax3.plot(x, z, marker='o', color="orange", alpha=0.5)

plt.show()
````

#### Boxplot - To show distributions with respect to categories.

A **box plot** (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.

The box plot allows quick graphical examination your data: 
- Box and whisker shows **[quartiles](https://en.wikipedia.org/wiki/Quartile)**
- The band inside the box is the **second quartile (the median)**
- The ends of the whiskers represents **minimum and maximum of all data** (most of the time - sometimes ends of the whisker can represent other values)

#### Violinplot

A **violin plot** is more informative than a box plot : 
- a box plot only shows summary statistics such as mean/median and interquartile ranges ; 
- a violin plot shows **the full distribution of the data**. 

‚û°Ô∏è The difference is particularly useful when the data distribution is multimodal (more than one peak). In this case a violin plot shows the presence of different peaks, their position and relative amplitude.

> üìö **Resources** : For a complete understanding of the violinplot chart, read [Wikipedia - Violin Plot](https://en.wikipedia.org/wiki/Violin_plot)

````python
# Example : `sns.violinplot` for distribution

plt.figure(figsize=(12, 6))

# Seaborn also has utility functions to load ready-to-play datasets
# This one concerns different tips in restaurants üí∏
df = sns.load_dataset("tips")

sns.violinplot(x="tip", data=df)

plt.title('Violin plot of tip variable')
plt.show()
````

````python
# `sns.violinplot` allows you to easily compare distributions

plt.figure(figsize=(12, 6))

sns.violinplot(x="time", y='tip', hue='sex', data=df, split=True)

plt.title('Violin plot of tip variable, depending on the time and the sex')
plt.show()
````

#### Histogram plot - To show univariate distribution of observations

The seaborn function `sns.distplot` combines the matplotlib `hist` function (with automatic calculation of a good default bin size) with the seaborn `kdeplot()` (that plots a smooth density estimation).

````python
# Example: `sns.displot` for distribution

plt.figure(figsize=(12, 6))

# Seaborn also has utility functions to load ready-to-play datasets
# This one concerns different types of flowers üå∫
df = sns.load_dataset("iris")

sns.distplot(df["sepal_length"])

plt.title('Distribution of Sepal Length')
plt.show()
````

#### Scatter plot - To plot a collection of data points

`sns.scatterplot` function draws a scatter plot of the observations. It displays values for typically two variables for a set of data. You can also display more variables by coloring the points, or varying the size.

````python
# Example: `sns.scatterplot` - for plotting data points

plt.figure(figsize=(12, 6))

sns.scatterplot(x="sepal_length", y="sepal_width", data=df, hue="species")

plt.show()
````

#### Pair plot - a magic function üé©

`sns.pairplot` is quite magical, you call it on your dataset and it draws **pairwise relationships in a dataset**. This means:
- each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column - `scatter_plot`
- The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column - `kde_plot`

````python
# Example: `sns.pairplot` - a magic function

# We specify another style to our seaborn viz
sns.set(style="ticks")

# Magic `pairplot` function
sns.pairplot(df, hue="species")

plt.show()
````

#### Adding one more dimention to your plot with sns.relplot and sns.catplot 

##### Relplot - To plot relations between continuous variables

Another useful seaborn function: `sns.relplot` for plotting 2 faces relational plots sharing the same X-Y axis and legend but with a varying column defined in `col`.

You can plot either :
- **Scatterplots**, by specifying the parameter `kind="scatter"`
- **Lineplots**, by specifying the parameter `kind="line"`.

````python
# Another example: `sns.relplot` - facing scatter plots

# Optional: we apply the default seaborn theme, scaling, and color palette
sns.set()

# Seaborn also has utility functions to load ready-to-play datasets
# This one concerns tips in restaurant üí∏
df = sns.load_dataset("tips")

plt.figure()

sns.relplot(x="total_bill", y="tip", hue="sex", col='time', data=df, kind="scatter")

plt.show()
````

##### Catplot - To plot comparisons between categorical variables

`sns.catplot` is used for plotting 2 faces categorical plots sharing the same X-Y axis and legend but with a varying column defined in `col`.

You can plot :
- **Boxplots**, by specifying the parameter `kind="box"`
- **Violinplots**, by specifying the parameter `kind="violin"`.
- **Barplots**, by specifying the parameter `kind="bar"`
- **Countplots**, by specifying the parameter `kind="count"`.
- And many others : refer to the [documentation](https://seaborn.pydata.org/generated/seaborn.catplot.html) for more details.

````python
# Another example: `sns.catplot` - facing box plots
plt.figure()

sns.catplot(x="sex", y="tip", hue="smoker", col='time', data=df, kind="box")

plt.show()
````

### III.1.C. Bokeh

[`bokeh`](https://bokeh.pydata.org/en/latest/) is another Python visualization library with some interactivity features.

For example: [Iris visualization](https://bokeh.pydata.org/en/latest/docs/gallery/iris.html) üå∫

````python
# Code is quite not so different from what we have seen before.
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.iris import flowers

colormap = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
colors = [colormap[x] for x in flowers['species']]

p = figure(title = "Iris Morphology")
p.xaxis.axis_label = 'Petal Length'
p.yaxis.axis_label = 'Petal Width'

p.circle(flowers["petal_length"], flowers["petal_width"],
         color=colors, fill_alpha=0.2, size=10)

show(p)
````

# II. Data cleaning

Before going further into the analysis, we need to check potential issues such as:
- missing or N/A values
- duplicated lines
- data not properly loaded
- etc.

and clean them.

### II.1. Handle Missing values

We can check missing values using `data.isnull().sum()`

- `data.isnull()` returns the DataFrame with all values replaced by a **boolean** indicating if the value is N/A
- we just have to apply `sum` method to the column axis of this DataFrame (as by default, the sum of an empty or NA value counts as 0)

````python
# Find missing data
data.isnull().sum()
````

We can solve those issues differently:
- A lot of missing values in Cabin column, the column does not appear useful: we will **drop the column**
- Only 2 in Embarked column, we will **drop the 2 rows**
- 177 in Age which should be an important feature, we will **replace empty values with the `mean()` of the existing values**

````python
# Cleaning Cabin - removing column
data_cleaned = data.drop(['Cabin'], axis=1)

# Cleaning Embarked - removing rows with missing values
data_cleaned = data_cleaned.dropna(subset=["Embarked"])

# Cleaning Age - replacing by mean value
data_cleaned["Age"] = data["Age"].fillna(data["Age"].mean())
````

### II.2. Remove Duplicated lines

Is there any duplicated lines? We can check using `duplicated()` method.

````python
# Find duplicated lines
data.duplicated().sum() # Returns 0 => there are no duplicated lines in the dataset
````

# III. Univariate analysis

Let's go deeper in our analysis. Especially let's try to understand **how passengers' characteristics impact their propability of survival**.

> **Remember that the quality of your analysis will depend on the questions you choose to answer.**

When performing data analysis, we start by exploring one variable and its distribution (univariate analysis), and then we can add up other variables and perform multivariate analysis.

## III.1. Gender

First, let's visualize the **number of passengers per gender** (univariate analysis).

````python
plt.figure(figsize=(6, 5))

# Plot the number of passengers per gender
sns.countplot(x='Sex',
              order=["female", "male"], 
              data=data_cleaned,
              palette="Blues")

# Write the labels on top
grouped_values = data_cleaned.groupby("Sex").Sex.count()
for label, value in enumerate(grouped_values):
    plt.text(label, value - 50, round(value,2), color='black', ha="center")

plt.title('Number of passengers per gender')
plt.show()
````

Now, let's visualize the **survival rate per gender**

````python
plt.figure(figsize=(6, 5))

# Plot the Survival Rate per Sex
sns.barplot(x='Sex',
            y='Survived',
            order=["female", "male"],
            data=data_cleaned,
            ci=None, # Removing confidence intervals
            palette="Blues")

# Write the labels on top
grouped_values = data_cleaned.groupby("Sex").Survived.mean()
for idx, value in enumerate(grouped_values):
    plt.text(idx, value - 0.1, round(value,2), color='black', ha="center")
    
plt.title('Survival rate per Gender')
plt.show()
````

We observe a **drastic difference in survival rate**! **74% of women survived, vs 19% of men**...

It appears that women were more likely to survive, but it might be too soon to conclude anything. Maybe another variable is the root cause behind these discrepancies.

Let's now analyse the variable Age and its relationship with the survival rate.

## III.2. Age

Again, let's start by visualizing the **distribution of our data** (univariate analysis).

> üî¶ **Hint**: We are visualizing `data` and not `data_cleaned` as we would have a peak at 30 - the mean value we used to replace NaN values

````python
plt.figure(figsize=(8,6))
sns.distplot(data.loc[data['Age'].notnull(), 'Age'])
plt.title('Age Distribution')
plt.show()
````

Does this distribution change depending on the Gender?

````python
plt.figure(figsize=(8,6))

sns.distplot(data.loc[data['Age'].notnull() & (data['Sex'] == 'male'), 'Age'],
             label='male')
sns.distplot(data.loc[data['Age'].notnull() & (data['Sex'] == 'female'), 'Age'],
             label='female')
plt.title('Age Distribution by Gender')
plt.legend()
plt.show()
````

Now let's explore the Survival rate depending on the Age.

````python
plt.figure(figsize=(8,6))

sns.distplot(data.loc[data['Age'].notnull() & (data['Survived'] == 1), 'Age'],
             label='Survived')
sns.distplot(data.loc[data['Age'].notnull() & (data['Survived'] == 0), 'Age'],
             label='Perished')
plt.title('Age Distribution by Survived')
plt.legend()
plt.show()
````

Not very "readable". We observe a peak of survivals around < 5 years old, but not sure how to interpret it.

Sounds more interesting to visualize survival rate depending on Age.

Let's think about a clever way on how to do that.

````python
def get_age_group(x):
    idx_group = int(x // 10)
    name_group = "{} to {}".format(idx_group*10, (idx_group+1)*10)
    return name_group

data_cleaned["Age_group"] = data_cleaned["Age"].apply(get_age_group)

sr_by_age_group = data_cleaned.groupby("Age_group")["Survived"].mean()
print(sr_by_age_group)
````

````python
plt.figure(figsize=(12, 8))

sns.barplot(x="Age_group", y="Survived", palette ="Blues", ci=None,
            order=sr_by_age_group.index[:-2],
            data=data_cleaned.loc[data_cleaned.Age < 70])
plt.show()
````

We can extend this analysis and break it per gender using `hue`

````python
plt.figure(figsize=(12, 8))

sns.barplot(x="Age_group", y="Survived", palette ="Blues", ci=None,
            hue="Sex",
            order=sr_by_age_group.index[:-2],
            data=data_cleaned.loc[data_cleaned.Age < 70])
plt.show()
````

Sounds safe to conclude: **Women and children first!** üö¢
